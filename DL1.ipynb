```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
le=LabelEncoder()
df['species']=le.fit_transform(df['species'])
df.head()
```

```python
df.info()
```

```python
df['species'].unique()
```

```python
X=df.drop('species',axis=1)
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
```

```python
sns.pairplot(df, hue='species')
```

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]

print(y)
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
```

```python
p.fit(x,y)
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values,y.values,clf=p,legend=2)
```

```python
from sklearn.datasets import make_classification
x_dt,y_dt=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,class_sep=0.5)

plt.figure(figsize=(10,6))
plt.scatter(x_dt[:,0],x_dt[:,1],c=y_dt,cmap='winter',s=100)
```
